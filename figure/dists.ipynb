```python
import mplfonts
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
exp = 'dists'
groups = [
    ('2nn', 'mnist'),
    ('cnn', 'fmnist'),
    ('lenet5', 'cifar10'),
    ('resnet9', 'cifar100'),
]
def path(exp, dataset, model, metric, index):
    return f'../results/{exp}-{dataset}-{model}/{metric}{index}.csv'
```

aquire data.

```python
for model, dataset in groups:
    for metric in ['acc']:
        dfs = [pd.read_csv(path(exp, dataset, model, metric, i), index_col=0) for i in range(5)]
        df = pd.concat(dfs, axis=0)
        
        df_mean = df.groupby(level=0).mean()
        print(df_mean)
        
        plt.plot()
        # 画折线图
        for i, (index, column) in enumerate(df_mean.items()):
            plt.plot(column.index, column.values, color=plt.cm.get_cmap('tab10')(i), label=str(index))
            plt.scatter(column.index, column.values, color=plt.cm.get_cmap('tab10')(i), s=10)  # 在每个点上画一个点
            
    plt.xlabel('Global Rounds')
    plt.ylabel('Accuracy')
    plt.legend()  # 显示图例
    plt.title(f'{dataset}'.capitalize())
    
    plt.savefig(f'{exp}-{dataset}.svg')
    plt.show()
```
